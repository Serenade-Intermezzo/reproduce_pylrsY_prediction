#Final prediction step

Due to the equatorial conformation showing better predictive results, we output the results of all equatorial conformations and select those for experimental verification based on probability.

```python
!/usr/local/envs/py38/bin/python -c "from autogluon.tabular import TabularDataset, TabularPredictor;\
import pandas as pd;\
import pickle;\
predictor = TabularPredictor.load('AutogluonModels/ag-20220629_024330/');\
f = open('reference_data/106_TCOC_SSM_evo256.pickle', 'rb');\
X_test, y_test = pickle.load(f);\
testset = pd.DataFrame(X_test);\
predictions = predictor.predict_proba(testset);\
f0 = open('reference_data/B1_ssm_tags.pickle', 'rb');\
tags = pickle.load(f0);\
df = pd.concat([predictions, tags], axis=1);\
df['rank']=df[True].rank(axis=0, method='min', ascending=False);\
print('Axial type:');\
print(df[0:171][df[0:171]['rank']<=20]);\
print('Equatorial type:');\
print(df[171:][df[171:]['rank']<=20]);\
df[171:].to_csv('equatorial_type_TCOC.csv', index=False);\
"
```
